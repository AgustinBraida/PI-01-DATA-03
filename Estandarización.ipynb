En los pasos posteriores, vamos a importar los archivos json y los pasamos a csv para luego exportarlos y poder cargarlos a la base de datos.

```python
constructors = pd.read_json(r'Datasets\constructors.json', lines=True)
results = pd.read_json(r'Datasets\results.json', lines=True)
drivers = pd.read_json(r'Datasets\drivers.json', lines=True)
```

```python
drivers.head()
```

```python
# Aca lo que se hizo fue pasar a str la columna que era un dicionario.
for i in range(len(drivers['name'])):
    drivers['name'][i] = drivers['name'][i]['forename'] + ' ' + drivers['name'][i]['surname']
```
